```python
data.rename(columns={"Unnamed: 0":"patient_id"},inplace=True)
```

```python
data.income = (data.income*100000).astype(int)
```

```python
data.age = [int(str(val).split(".")[1]) for val in data.age]
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.figure(figsize=(5, 4))
sns.countplot(x=data.gender)
plt.ylabel('No. of patients')
plt.title('Total count of patients by gender')
plt.show()
```

```python
plt.figure(figsize=(11, 4))
sns.countplot(hue=data.gender, x=data.age)
plt.ylabel('No. of patients')
plt.title('Patients by Age (separated by gender)')
plt.show()
```

```python
plt.figure(figsize=(12, 4))
sns.barplot(x='age', y='visits', data=data)
plt.title("Average Doctor Visits by Age")
plt.xlabel("Age")
plt.ylabel("Average Visits")
plt.show()
```

```python
plt.figure(figsize=(5, 4))
sns.countplot(x=data.private)
plt.ylabel('No. of patients')
plt.title('Patients with Private Health Insurance')
plt.show()
```
